```python
# Dependencies and Setup mike
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import gmaps
import os
import scipy.stats as sts
from scipy.stats import linregress

# Import API key
from api_keys import g_key
```

```python
# Name of the CSV file
top_20 = "Resources/top_20_ca_wildfires.csv"
ca_rainfall_df = "Resources/ca_rainfall_df.csv"
```

```python
# read the CSV in pandas
top_20 = pd.read_csv(top_20)
top_20
```

```python
index = top_20["index"]
fire_name = top_20["fire_name"]
cause = top_20["cause"]
month = top_20["month"]
year = top_20["year"]
county = top_20["county"]
acres = top_20["acres"]
structures = top_20["structures"]
deaths = top_20["deaths"]
```

```python
# Rename column year 
summary_top20_df = pd.DataFrame({"Index": index, "Fire Name": fire_name, "Cause": cause, "Month": month, 
                                 "Year": year, "County": county, "Acres": acres, 
                                 "Structures": structures, "Deaths": deaths})

summary_top20_df
```

```python
# read the CSV in pandas
ca_rainfall_df = pd.read_csv(ca_rainfall_df)
ca_rainfall_df
```

```python
#Merge on year
merge_df = pd.merge(summary_top20_df, ca_rainfall_df, on="Index", left_index=True)
merge_df
```

```python
merge_drop_df = merge_df.drop(['County_x','Month_x', 'Year_x'], axis = 1) 
merge_drop_df
```

```python
# rename
final_df = merge_drop_df.rename(columns={"County_y": "County", "Month_y": "Month", "Year_y": "Year"})
final_df
```

```python
# create output csv
final_df.to_csv("Resources/final_df.csv")
```

```python
# Is there a correlation between total number of acres burned and rainfall?
# Summary Stats
mean_numpy = np.mean(final_df["Acres"])
print(f"The mean Acres burned is {mean_numpy}")
median_numpy = np.median(final_df["Acres"])
print(f"The median Acres burned is {median_numpy}")
mode_scipy = sts.mode(final_df["Acres"])
print(f"The mode Acres burned is {mode_scipy}")
print("----------------------------------------")
mean_numpy = np.mean(final_df["Rainfall (in)"])
print(f"The mean Rainfall (in) is {mean_numpy}")
median_numpy = np.median(final_df["Rainfall (in)"])
print(f"The median Rainfall (in) is {median_numpy}")
mode_scipy = sts.mode(final_df["Rainfall (in)"])
print(f"The mode Rainfall (in) is {mode_scipy}")
```

```python
# Histogram of the Acres burned data
# Normal test of the data
plt.hist(final_df["Acres"])
plt.xlabel('Acres Burned')
plt.ylabel('Counts')
plt.show()
print(sts.normaltest(final_df["Acres"]))
```

```python
# Histogram of the Rainfall (in) data
# Normal test of the data
plt.hist(final_df["Rainfall (in)"])
plt.xlabel('Rainfall (in)')
plt.ylabel('Counts')
plt.show()
print(sts.normaltest(final_df["Rainfall (in)"]))
```

```python
# Scatter Plot
plt.scatter(final_df["Acres"],final_df["Rainfall (in)"])
clb = plt.colorbar()
plt.xlabel("Acres Burned")
plt.ylabel("Rainfall (in)")
clb.set_label("Rainfall Compared to Acres Burned")
plt.show()
```

```python
acres_burned = final_df["Acres"]
rainfall = final_df["Rainfall (in)"]
```

```python
# Calculate the correlation coefficient between Acres Burned and Rainfall (in)
# A coefficient of 0.07 which is near 0 shows that Acres Burned appears unrelated to Rainfall (in)
print(f"The correlation coefficient between Acres Burned and Rainfall (in) is {round(st.pearsonr(acres_burned,rainfall)[0],2)}")
```

```python
# Remove the outlier, Zaca wildfire in 2018 and redo scatter
mod_df = final_df.drop([5])
mod_df
```

```python
# Scatter Plot
plt.scatter(mod_df["Acres"],mod_df["Rainfall (in)"])
clb = plt.colorbar()
plt.xlabel("Acres Burned")
plt.ylabel("Rainfall (in)")
clb.set_label("Rainfall Compared to Acres Burned")
plt.show()
```

```python
acres_burned_mod = mod_df["Acres"]
rainfall_mod = mod_df["Rainfall (in)"]
```

```python
# Calculate the correlation coefficient between Acres Burned and Rainfall (in) with Zaca dropped as an outlier
# A coefficient of -0.15 is still near 0 but does show a slightly stronger correlation trend negative after removal of
# the Zaca wildfire of 2018
print(f"The correlation coefficient between Acres Burned and Rainfall (in) is {round(st.pearsonr(acres_burned_mod,rainfall_mod)[0],2)}")
```

```python
# Add the linear regression equation and line to plot using the mod_df with removal of the Zaca wildfire
# As stated above the correlation is not a good fit, a value above at least .50 would allow us to confirm that there
# is a relationship
x_values = mod_df['Acres']
y_values = mod_df['Rainfall (in)']
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.scatter(x_values,y_values)
plt.plot(x_values,regress_values,"r-")
plt.xlabel('Acres Burned')
plt.ylabel('Rainfall(in)')
print(f"The r-squared is: {rvalue**2}")
plt.show()
```
